# KMeans clustering

```python
from sklearn.cluster import KMeans
inertia_values = []

cluster_range = range(1, 11)
for k in cluster_range:
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    model.fit(df)
    inertia_values.append(model.inertia_)


plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow method for optimal')
plt.xticks(cluster_range)
plt.show()
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(df.drop(columns=['Cluster']), df['Cluster'])
print("Silhouette Score:", (sil_score + 1) / 2)
```

```python
df.Cluster.value_counts()
```

# Decomposition without scaled

```python
def visualize_dim_red(name, alg, data, clusters):
    al = alg(n_components=2)
    tr = al.fit_transform(data)

    plt.scatter(tr[:, 0], tr[:, 1], c=clusters, cmap='viridis', alpha=0.7, edgecolors='k')

    plt.title(name)
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.colorbar(label="Cluster Label")
    plt.show()

algs = {'PCA': PCA, 'tSNE': TSNE, 'UMAP': UMAP, 'ICA': FastICA, 'Isomap': Isomap}

 
clusters = df['Cluster']


df_features = df.drop(columns=['Cluster'])


for name, alg in algs.items():
    visualize_dim_red(name, alg, df_features, clusters)
```

# Scaled data

```python
from sklearn.preprocessing import StandardScaler
columns = df.columns
df_scaled = df.copy()
scaler = StandardScaler()
df_scaled[columns] = scaler.fit_transform(df[columns])
```

```python
df_scaled
```

# Kmeans clustering

```python
inertia_values = []

cluster_range = range(1, 11)
for k in cluster_range:
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    model.fit(df_scaled)
    inertia_values.append(model.inertia_)


plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow method for optimal')
plt.xticks(cluster_range)
plt.show()
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_scaled['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(df_scaled.drop(columns=['Cluster']), df_scaled['Cluster'])
print("Silhouette Score:", (sil_score + 1) / 2)
```

```python
df_scaled.Cluster.value_counts()
```

# Decomposition with Scale

```python

def visualize_dim_red(name, alg, data, clusters):
    al = alg(n_components=2)
    tr = al.fit_transform(data)
    plt.scatter(tr[:, 0], tr[:, 1], c=clusters, cmap='viridis', alpha=0.7, edgecolors='k')
    plt.title(name)
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.colorbar(label="Cluster Label")
    plt.show()

algs = {'PCA': PCA, 'tSNE': TSNE, 'UMAP':UMAP, 'ICA': FastICA, 'Isomap': Isomap}

clusters = df_scaled['Cluster']
df_features = df_scaled.drop(columns=['Cluster'])

for name, alg in algs.items():
    visualize_dim_red(name, alg, df_features, clusters)
```
